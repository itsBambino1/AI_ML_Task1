```python
from mistralai.client import Mistral
```

```python
api_key = "IdV7UeOvTTFu0IZzEf8ciFMMuCKr9eL5" 
client = Mistral(api_key=api_key) # Initialize the Mistral client with the API key
```

```python
SYSTEM_PROMPT = """
You are a helpful medical assistant chatbot.
Rules:
1. Provide general health information only.
2. Be friendly, simple, and easy to understand.
3. Do not diagnose medical conditions.
4. Do not prescribe medications or dosages.
5. Encourage doctor consultation for serious symptoms.
6. Refuse dangerous, harmful, or emergency medical advice requests.
7. Mention that your advice is informational only.
""" # Initialize the system prompt for the medical assistant chatbot
```

```python
dangerous_keywords = [
    "suicide",
    "kill myself",
    "self harm",
    "overdose",
    "poison",
    "harm myself",
    "die",
    "emergency",
    "chest pain",
    "heart attack",
    "stroke",
    "can i take 20 tablets",
    "medicine overdose",
    "stop breathing",
    "severe bleeding",
    "fainting",
    "unconscious"
] # List of dangerous keywords for medical advice requests
```

```python
def safety_filter(query): # Filter out potentially dangerous queries
    query = query.lower()

    for keyword in dangerous_keywords:
        if keyword in query:
            return False

    return True 
```

```python
def health_chatbot(user_query): # Handle user queries and provide responses
    
    if not safety_filter(user_query):
        return """
This question may involve a medical emergency or harmful situation.

Please seek immediate help:
- Contact emergency services
- Visit nearest hospital
- Talk to a licensed doctor
"""

    response = client.chat.complete(
        model="mistral-large-latest",
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": user_query}
        ]
    )

    return response.choices[0].message.content
```

```python
query = "What causes sore throat?"
print(health_chatbot(query))
```

```python
query = "I feel like killing myself, what should I do?"
print(health_chatbot(query))
```

```python
query = "I think I am having a heart attack"
print(health_chatbot(query))
```

```python
while True: 
    query = input("Ask health question (or type exit): ")

    if query.lower() == "exit":
        break

    response = health_chatbot(query)
    print("\nBot:", response)
    print()
```
